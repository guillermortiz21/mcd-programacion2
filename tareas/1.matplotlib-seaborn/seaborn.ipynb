# Seaborn

```python
# Importar librer√≠a
import seaborn as sns
```

```python
# Cargar dataset y mostrarlo
df = sns.load_dataset("tips")
df.head()
```

```python
# Gr√°ficos de barras
sns.countplot(x='sex', data=df)
```

```python
# gr√°fico de barras
sns.countplot(x="day", data=df)
```

```python
# Diagramas de barras
sns.barplot(x="sex", y="total_bill", data=df)
```

```python
# Diagrama de caja y bigotes
sns.boxplot(data=df, x="sex", y= "total_bill")
```

```python
# Grafico de regresi√≥n

sns.lmplot(x="total_bill", y="tip", data=df)
```

```python
# An√°lisis para varias variables

sns.pairplot(df)
```

```python
# Customizaciones

sns.set_theme()

tips = sns.load_dataset("tips")

# Crear visualizac√≠on
sns.relplot(
    data=tips,
    x = "total_bill",
    y = "tip",
    col = "time",
    hue = "smoker",
    style = "smoker",
    size = "size"
)
```

```python
# Relplot

dots = sns.load_dataset("dots")
sns.relplot(
    data = dots, kind = "line", x = "time", y = "firing_rate",
    col = "align", hue = "choice", size = "coherence", style = "choice",
    facet_kws = dict(sharex=False)
)
```

```python
# Estimaciones estad√≠sticas

fmri = sns.load_dataset("fmri")

sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal",
    col="region", hue="event", style="event"
)
```

```python
sns.lmplot(data=tips, x="total_bill",y="tip", col="time", hue="smoker")
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
penguins = sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="flipper_length_mm",
              y="bill_length_mm", hue="species")
```

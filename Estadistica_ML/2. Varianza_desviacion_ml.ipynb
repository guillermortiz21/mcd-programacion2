# Medidas de Varianza y Desviación Estándar Aplicadas a Machine Learning

En este notebook se abordan dos de las medidas de dispersión más importantes en el análisis de datos y Machine Learning: **la varianza** y **la desviación estándar**. Estas medidas permiten conocer la dispersión de los datos en torno a la media, lo que es fundamental para entender la variabilidad en un conjunto de datos y para el preprocesamiento de los mismos en modelos predictivos.

A lo largo del notebook se presentan:

- **Conceptos teóricos** con fórmulas y descripción de cada variable.
- **Tres ejemplos prácticos** para el cálculo de la varianza y la desviación estándar usando Python.
- Un **ejemplo real** utilizando el dataset *Iris* para profundizar en la aplicación de estas medidas y su uso en la normalización de datos.

Recordemos las fórmulas:

### Varianza

$$ \sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2 $$

### Desviación Estándar

$$ \sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2} $$

donde:

- **\( n \)** es el número de observaciones.
- **\( x_i \)** es el valor de la i-ésima observación.
- **\( \bar{x} \)** es la media de los valores.

## 1. Varianza

La **varianza** es una medida que cuantifica la dispersión de los datos respecto a la media. Se define como el promedio de los cuadrados de las diferencias entre cada valor y la media del conjunto de datos.

La fórmula de la varianza es:

$$ \sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2 $$

donde:

- **\( n \)** es el número de observaciones en el conjunto de datos.
- **\( x_i \)** es el valor de la i-ésima observación.
- **\( \bar{x} \)** es la media de los valores.

En el contexto de Machine Learning, la varianza es útil para conocer la dispersión de las características de los datos y para identificar posibles valores atípicos.

### Ejemplos de Cálculo de la Varianza

A continuación se presentan **tres ejemplos** de cálculo de la varianza utilizando Python:

```python
import numpy as np

# La varianza se puede calcular con numpy: np.var()

# Ejemplo 1: Varianza de un arreglo simple
datos1 = np.array([10, 20, 30, 40, 50])
varianza1 = np.var(datos1)
print('Ejemplo 1 - Varianza:', varianza1)

# Ejemplo 2: Varianza de un conjunto de edades
edades = np.array([22, 25, 25, 30, 34, 40, 40, 40, 45])
varianza_edades = np.var(edades)
print('Ejemplo 2 - Varianza de edades:', round(varianza_edades,2))

# Ejemplo 3: Varianza de puntuaciones en un examen
puntuaciones = np.array([85, 90, 78, 92, 88, 76, 95])
varianza_puntuaciones = np.var(puntuaciones)
print('Ejemplo 3 - Varianza de puntuaciones:', round(varianza_puntuaciones,2))
```

## 2. Desviación Estándar

La **desviación estándar** es la raíz cuadrada de la varianza y se expresa en las mismas unidades que los datos originales. Esta medida es especialmente útil para interpretar la dispersión de los datos de manera intuitiva.

La fórmula de la desviación estándar es:

$$ \sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2} $$

donde los símbolos representan lo mismo que en la fórmula de la varianza.

En Machine Learning, la desviación estándar se utiliza en la normalización y escalado de datos, fundamentales para el correcto funcionamiento de muchos algoritmos.

### Ejemplos de Cálculo de la Desviación Estándar

A continuación se presentan **tres ejemplos** de cálculo de la desviación estándar utilizando Python:

```python
import numpy as np

# La desviación estándar se puede calcular con numpy: np.std()

# Ejemplo 1: Desviación estándar de un arreglo simple
datos1 = np.array([10, 20, 30, 40, 50])
desviacion1 = np.std(datos1)
print('Ejemplo 1 - Desviación estándar:', round(desviacion1,2))

# Ejemplo 2: Desviación estándar de un conjunto de edades
edades = np.array([22, 25, 25, 30, 34, 40, 40, 40, 45])
desviacion_edades = np.std(edades)
print('Ejemplo 2 - Desviación estándar de edades:', round(desviacion_edades,2))

# Ejemplo 3: Desviación estándar de puntuaciones en un examen
puntuaciones = np.array([85, 90, 78, 92, 88, 76, 95])
desviacion_puntuaciones = np.std(puntuaciones)
print('Ejemplo 3 - Desviación estándar de puntuaciones:', round(desviacion_puntuaciones,2))
```

## 3. Ejemplo Real: Aplicación en el Dataset Iris

En esta sección se utiliza el dataset **Iris** para demostrar cómo se calculan y aplican la varianza y la desviación estándar en un escenario real. Se procede a:

- Cargar el dataset Iris.
- Calcular la varianza y la desviación estándar para cada una de las características.
- Visualizar la dispersión de los datos mediante histogramas y un heatmap de la matriz de correlación.
- Realizar la **normalización (estandarización)** de los datos usando la media y la desviación estándar.

Recordemos la fórmula para la estandarización:

$$ z = \frac{x - \bar{x}}{\sigma} $$

donde \( x \) es el valor original, \( \bar{x} \) es la media y \( \sigma \) es la desviación estándar.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

# Cargar el dataset Iris
iris = datasets.load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target

print('Primeras 5 filas del dataset Iris:')
display(df_iris.head())

print('\nDescripción estadística del dataset:')
display(df_iris.describe()) # Obtiene estadística como media, desviación estándar
# cuartiles, máximos y mínimos.
```

```python
# Calcular varianza y desviación estándar para cada característica
features = iris.feature_names

print('Medidas de dispersión para cada característica:')
for feature in features:
    valores = df_iris[feature]
    media = np.mean(valores)
    varianza = np.var(valores)
    desviacion = np.std(valores)
    
    print(f"\nCaracterística: {feature}")
    print(f"Media: {media:.4f}") #:.4f es para mostrar el resultado con n decimales
    print(f"Varianza: {varianza:.4f}")
    print(f"Desviación Estándar: {desviacion:.4f}")
```

```python
# Visualización de la dispersión

# Histograma para la longitud del sépalo
plt.figure(figsize=(8, 4))
sns.histplot(df_iris['sepal length (cm)'], kde=True, bins=20, color='skyblue')
plt.title('Histograma: Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frecuencia')
plt.show()

# Histograma para el ancho del sépalo
plt.figure(figsize=(8, 4))
sns.histplot(df_iris['sepal width (cm)'], kde=True, bins=20, color='skyblue')
plt.title('Histograma: Sepal width')
plt.xlabel('Sepal width (cm)')
plt.ylabel('Frecuencia')
plt.show()

# Heatmap de la matriz de correlación
plt.figure(figsize=(8, 6))
corr = df_iris[features].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Matriz de correlación de las características')
plt.show()
```

## 4. Normalización (Estandarización) de las Características

La normalización de los datos es un paso fundamental en el preprocesamiento en Machine Learning. Utilizaremos la **estandarización** para transformar las características del dataset Iris utilizando la fórmula:

$$ z = \frac{x - \bar{x}}{\sigma} $$

A continuación se muestra cómo aplicar esta transformación.

```python
from sklearn.preprocessing import StandardScaler

# Seleccionamos solo las características numéricas
X = df_iris[features].values

# Inicializamos el escalador
scaler = StandardScaler()

# Ajustamos y transformamos los datos
X_scaled = scaler.fit_transform(X)

# Se mueven los datos a estar centrados en cero con una desviación estándar
# de 1

# Convertimos los datos escalados en un DataFrame
df_iris_scaled = pd.DataFrame(X_scaled, columns=features)
df_iris_scaled['target'] = df_iris['target']

print('Primeras 5 filas del dataset normalizado:')
print(df_iris_scaled.head())

# Comparación: Distribución de 'sepal length (cm)' antes y después de la normalización
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_iris['sepal length (cm)'], kde=True, bins=20, color='skyblue')
plt.title('Antes de la normalización')

plt.subplot(1, 2, 2)
sns.histplot(df_iris_scaled['sepal length (cm)'], kde=True, bins=20, color='salmon')
plt.title('Después de la normalización')

plt.tight_layout()
plt.show()
```
